### extract_technos

```python
def extract_technos(text):
    words = re.split(r'\W+', text)
    technos = {w for w in words if w.lower() in TECHNOS}
    return list(technos)
```

### add_technos_to_stack

```python
preprocessed['stack'] = preprocessed['text'].apply(lambda x: extract_technos(x))
```

### pivot_technos

```python
def pivot_technos(df):
    """ 
    Melt dataframe to have one technology per row (for usage in Tableau). 
    """
    unpivotted_columns = ['id', 'url', 'title', 'company', 'location', 'type', 'industry', 'remote', 'created_at',
                          'text', 'stack']
    pivotted_technos = pd.melt(df, id_vars=unpivotted_columns).sort_values(by=['company', 'created_at', 'title'])
    pivotted_technos.reset_index(drop=True, inplace=True)
    return pivotted_technos
```

### clean_df

```python
def clean_pivot(df):
    df['technos'] = df['value']
    df.drop(['variable', 'stack', 'value'], axis=1, inplace=True)
    df.dropna(subset='technos', inplace=True)
    return df
```
